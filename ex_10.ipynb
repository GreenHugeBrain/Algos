<h2>ფეხბურთის სტატისტიკა</h2>
<br>
მოცემული გაქვთ base ფაილი , სადაც ასახულია საფეხბურთო გუნდების მიერ ჩატარებული თამაშების სტატისტიკა

<br>

მოგებების რაოდენობა : Wins 
<br>
წაგებების რაოდენობა : Losses 

ფრედ დასრულები თამაშების რაოდენობა : Draws 

გატანილი გოლების რაოდენობა : Goals 

თათვლილი გოლების რაოდენობა : Goals Allowed 

სარეიტინგო ქულა : Points 

------
• მონაცემები გააერთიანეთ მონაცემთა რომელიმე სტრუქტურაში **(გამოიყენეთ მონაცემთა ისეთი სტრუქტურა რომელიც დაგეხმარებათ ამოცანის პირობების შესრულებაში - სტატისტიკის მომზადებაში**
<br>

• ვითვლით თითოეული გუნდის მოგებების პროცენტულ მაჩვენებელს **(რამდენი პროცენტია მოგებების საერთოდ ჩატარებული თამაშებიდან)** და ცალკე ფაილში  ვინახავთ ყველაზე მაღალი პროცენტული მაჩვენებლით 3 საფეხბურთო გუნდს **(სამი საუკეთესო მოგების მაჩვენებლიანი გუნდი)**
<br>

• ასევე ცალკე ფაილში ვინახავთ იმ გუნდების ჩანაწერებს, რომელთა **წაგებული მატჩების რაოდენობა აღემატება როგორც მოგებულს ასევე ფრედ დასრულებულ თამაშს**. (ცუდ მაჩვენებლიანი გუნდები)
<br>

**! მონაცემებში არის ზოგან # ზოგან FLOAT, BOOL STRING მონაცემები და TRY EXCEPT გამოიყენეთ** რათა მონაცემების დალაგება სწორად შეძლოთ
